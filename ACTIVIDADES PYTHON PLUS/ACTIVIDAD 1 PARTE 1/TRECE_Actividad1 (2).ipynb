## Resolvé el siguiente desafío:


**EJERCICIO 13:** dado el archivo denominado [log_catedras](https://archivos.linti.unlp.edu.ar/index.php/s/ofgZh41dUYMYX6H) que contiene el registro de accesos al entorno catedras.linti, se desea:

1. Generar una función que retorne los usuarios que alguna vez vieron el estado de la entrega 1 desde la facultad. Esto se puede chequear si la columna "Contexto del evento" contiene el texto "Tarea: Entrega 1" y la columna "Nombre evento" contiene el texto "Se ha visualizado el estado de la entrega." y la dirección IP comienza con "163.10". Esta función recibe un parámetro opcional que puede contener: "A", para pasar los nombres de usuario a mayúsculas: "a", para pasarlos a mimúsculas o "None" si se dejan tal como se encuentran en el archivo. Por defecto la función retorna los nombres tales como aparecen en el archivo.
2. Escribir un programa que, utilizando la función anterior, muestre un listado similar al siguiente:

```
---------------------
Usuario en el sistema
---------------------

    Butterfree
    Pikachu
    Oddish
    Kingler
    Tauros
```

Ante igual cantidad de accesos, ordenar en forma alfabética ascendente.


````python
#Solución
import csv
import os

def conditions (user):
    """Esta función procesa un diccionario y devuelve True si un elemento cumple con algunas condiciones.
    
    Argumentos:
        user [dict]: es un diccionario creado a partir de csv.Dictreader.
        
    Retorno:
        [boolean]: True si el elemento cumple con algunas condiciones, de lo contrario False.
    """

    if ("Tarea: Entrega 1" in user["Contexto del evento"]):
        if ("Se ha visualizado el estado de la entrega." in user["Nombre evento"]):
            if(user["Dirección IP"].startswith("163.10")):
                return True
    return False

def generate_users (csvreader, Format = None):
    """Esta función procesa elementos del archivo csvreader y los convierte en una lista. 
       Luego, utiliza el Argument Format para convertir cada elemento de la lista en mayúsculas 
       o minúsculas, según corresponda.

    Argumentos:
        csvreader [_reader]: es un archivo con diccionarios.
        Format [str]: es la clave para convertir los elementos en mayúsculas o minúsculas.
    
    Retorno:
        [list]: una lista con el formato correcto
    """

    user_filter = list(filter(conditions, csvreader))
    user_names = sorted(list({user["Nombre completo del usuario"] for user in user_filter}))
    match Format:
        case "A":
            return [name.upper() for name in user_names]
        case "a":
            return [name.lower() for name in user_names]
        case _ :
            return user_names


from collections import Counter
path = os.path.realpath("")
file_path = os.path.join(path, "log_catedras.csv")
with open(file_path, 'r', encoding="utf-8") as file:
    csvreader = csv.DictReader (file, delimiter=",")
    header = next(csvreader)
    Format = input("Enter \"A\" to make the list in uppercase, or, enter \"a\" to make the list in lowercase.")
    users = generate_users(csvreader, Format)
print ("-" * 12)
print ("System Users")
print ("-" * 12)
[print (elem) for elem in users]
````
